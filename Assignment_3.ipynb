**Import required libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Read the dataset**

```python
df=pd.read_csv('/content/StudentsPerformance.csv')
```

**Find out how many males and females participated in the test**

```python
df[df.gender=='female'].gender.count()
```

518 females participated in the test

```python
df[df.gender=='male'].gender.count()
```

482 males participated in the test

**Student's parental level of education**

```python
df['parental level of education'].value_counts()
```

Among parents of the students. Parents with college level of education is prominent

**Who scores the most on average for math,reading and writing based on gender**

```python
df.groupby('gender')[['math score','reading score','writing score']].mean()
```

male has more  math score than female. female has more reading and writing  score than male.

**Who scores the most on average for math, reading and writing based on test preparation course**

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
```

math score,reading and writing score average is more to the students who completed test preparation course

**Scoring variation for math,reading and writing based on gender**

```python
df.groupby('gender')[['math score','reading score','writing score']].var()
```

Scoring variation of female in math score, reading and writing score is more than male

**Scoring variation for math,reading and writing based on test preparation course**

```python
df.groupby('test preparation course')[['math score','reading score','writing score']].var()
```

math score,reading and writing score variation  is more to the students who does not completed test preparation course

**Find out top 25% of students based on their math  score**

```python
np.quantile(df['math score'],0.75)
```
